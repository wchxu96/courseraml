### implementing k-means

```python
def computeDistance(point1,point2):
    assert point1.shape == point2.shape
    return np.sum(np.square(point1-point2))
```

```python
def findClosestCentroids(myX,centroids,K):
    '''
    myX is the data point(m * n ndarray) while the centroids is the 
    centroid (K * n ndarray)
    K is just an assertion 
    return a idx which indicate which group the myX points belong to
    (m * 1 ndaaray)
    '''
    assert K == centroids.shape[0]
    assert centroids.shape[1] == myX.shape[1]
    m = myX.shape[0]
    n = myX.shape[1]
    idx = np.zeros((m,1))
    for myX_index in xrange(m):
        idx[myX_index] = np.argmin(map(lambda x: computeDistance(myX[myX_index],x),centroids))
    return idx    
```

```python
print findClosestCentroids(X,initial_centroids,3)[1]
```

```python
def computeCentroids(myX,idx,K):
    '''
    compute the new centroids
    '''
    #assert np.unique()
    m = myX.shape[0]
    n = myX.shape[1]
    new_centroid = np.zeros((K,n))
    cluster_num = np.zeros((K,n))
    for item in zip(myX,idx):
        index = int(item[1][0])
        cluster_num[index] += 1
        new_centroid[index] += item[0]
    return new_centroid / cluster_num
```

```python
def kmeans(myX,initial_centroids,n_iter,K):
    centroids = initial_centroids
    for x_iter in xrange(n_iter):
        idx = findClosestCentroids(myX,centroids,K)
        centroids = computeCentroids(myX,idx,K)
    return idx,centroids
    
```

```python
idx,centroids = kmeans(X,initial_centroids,10,3)
#print centroids.shape
#print centroids[1,:]
```

```python
def get_slice(X,idx,K):
    X_Slice = [[] for _ in range(K)]
    for i in xrange(idx.shape[0]):
        item = idx[i]
        index = int(item[0])
        #print index
        X_Slice[index].append(X[i])
    return X_Slice
```

```python
#print X_Slice[0][0] == X_Slice[0][1]
def slice_X_y(X_Slice_item):
    tempx,tempy = [],[]
    for item in X_Slice_item:
        tempx.append(item[0])
        tempy.append(item[1])
    return tempx,tempy
```

```python
def plotData(X_Slice,centroids):
    scatter_3 = go.Scatter(
        x = slice_X_y(X_Slice[0])[0],
        y = slice_X_y(X_Slice[0])[1],
        mode = 'markers',
        marker = dict(symbol="star",size=8),
        name = 'cluster 1'
    )
    scatter_4 = go.Scatter(
        x = slice_X_y(X_Slice[1])[0],
        y = slice_X_y(X_Slice[1])[1],
        mode = 'markers',
        marker = dict(symbol="cross",size=8),
        name = 'cluster 2'
    )
    scatter_5 = go.Scatter(
        x = slice_X_y(X_Slice[2])[0],
        y = slice_X_y(X_Slice[2])[1],
        mode = 'markers',
        marker = dict(symbol="circle",size=8),
        name = 'cluster 3'
    )
    scatter_6 =go.Scatter(
        x = centroids[:,0],
        y = centroids[:,1],
        mode = 'markers',
        marker = dict(symbol="star-diamond",size=12),
        name = 'centroid'
    )

    data = [scatter_3,scatter_4,scatter_5,scatter_6]
    return py.iplot(data)
    
```

```python
#from IPython.display import Image,display
X_Slice = get_slice(X,idx,K)
plotData(X_Slice,centroids)
```

```python
def chooseKRandomCentroids(myX, K):
    rand_indices = sample(xrange(0,myX.shape[0]),K)
    return np.array([myX[i] for i in rand_indices])
```

### implementing PCA

```python
data_file2 = "./data/ex7data1.mat"
data = scipy.io.loadmat(data_file2)
X_pca = data['X']
X_pca.shape
```

```python
scatter_8 = go.Scatter(
    x = X_pca[:,0],
    y = X_pca[:,1],
    mode = "markers"
)
data = [scatter_8]
py.iplot(data)
```

```python
def normalize(X):
    '''
    normalizing the features before pca is important and neccesary.
    subtract the mean of each feature and scale by std dev.
    '''
    X_mean = np.mean(X,axis=0)
    X_std = np.std(X,axis=0)
    X = (X - X_mean) / X_std
    return X,X_mean,X_std
```

```python
X_pca,X_mean,X_std = normalize(X_pca)
```

```python
X_pca.shape
```

```python
def pca(X):
    m = X.shape[0]
    sigma = (1./ m) * np.dot(X.T,X)
    U,s,Vh = scipy.linalg.svd(sigma)
    #print U.shape,s.shape,Vh.shape
    return U,s,Vh
```

```python
U,s,Vh = pca(X_pca)
```

```python
###这里的绘图稍有疑问.
###PCA中特征向量是方差最大的方向
###但特征值代表什么呢？
###为什么这样画出来与andrew ng 给的图是一致的？
scatter_9 = go.Scatter(
    x = [X_mean[0],X_mean[0] + 1.5 * s[0] * U[0,0]],
    y = [X_mean[1],X_mean[1] + 1.5 * s[0] * U[0,1]],
    mode = 'lines',
    line = dict(color = 'red',shape='linear')
)
scatter_10 = go.Scatter(
    x = [X_mean[0],X_mean[0] + 1.5 * s[1] * U[1,0]],
    y = [X_mean[1],X_mean[1] + 1.5 * s[1] * U[1,1]],
    mode = 'lines',
    line = dict(color = 'green',shape='linear')
)
data = [scatter_8,scatter_9,scatter_10]
py.iplot(data)
```

```python
#这个是sklearn中的transform过程
def projectData(myX,myU,K):
    reducU = myU[:,:K]
    X_red = np.dot(myX,reducU)
    return X_red
```

```python
reduce_data = projectData(X_pca,U,1)
```

```python
print reduce_data[0]
```

```python
def recoverData(red_X,myU,K):
    '''
    a matrix can transform from A to B
    '''
    former_X = np.dot(red_X,myU[:,:K].T)
    return former_X
```

```python
recover_data = recoverData(reduce_data,U,1)
```

### 2.4 Face Image Dataset

```python
datafile_face = './data/ex7faces.mat'
data = scipy.io.loadmat(datafile_face)
X_face = data['X']
```

```python
X_face.shape
```

```python
def getDatumImg(row):
    height,weight = 32,32
    image = row.reshape(height,weight)
    return image.T

def displayData(myX, mynrows = 10, myncols = 10):
    """
    Function that picks the first 100 rows from X, creates an image from each,
    then stitches them together into a 10x10 grid of images, and shows it.
    """
    width, height = 32, 32
    nrows, ncols = mynrows, myncols

    big_picture = np.zeros((height*nrows,width*ncols))
    
    irow, icol = 0, 0
    for idx in xrange(nrows*ncols):
        if icol == ncols:
            irow += 1
            icol  = 0
        iimg = getDatumImg(myX[idx])
        big_picture[irow*height:irow*height+iimg.shape[0],icol*width:icol*width+iimg.shape[1]] = iimg
        icol += 1
    fig = plt.figure(figsize=(10,10))
    img = scipy.misc.toimage( big_picture )
    plt.imshow(img,cmap = cm.Greys_r)
```

```python
displayData(X_face)
```
