### implementing k-means

```python
def computeDistance(point1,point2):
    assert point1.shape == point2.shape
    return np.sum(np.square(point1-point2))
```

```python
def findClosestCentroids(myX,centroids,K):
    '''
    myX is the data point(m * n ndarray) while the centroids is the 
    centroid (K * n ndarray)
    K is just an assertion 
    return a idx which indicate which group the myX points belong to
    (m * 1 ndaaray)
    '''
    assert K == centroids.shape[0]
    assert centroids.shape[1] == myX.shape[1]
    m = myX.shape[0]
    n = myX.shape[1]
    idx = np.zeros((m,1))
    for myX_index in xrange(m):
        idx[myX_index] = np.argmin(map(lambda x: computeDistance(myX[myX_index],x),centroids))
    return idx    
```

```python
print findClosestCentroids(X,initial_centroids,3)[1]
```

```python
def computeCentroids(myX,idx,K):
    '''
    compute the new centroids
    '''
    #assert np.unique()
    m = myX.shape[0]
    n = myX.shape[1]
    new_centroid = np.zeros((K,n))
    cluster_num = np.zeros((K,n))
    for item in zip(myX,idx):
        index = int(item[1][0])
        cluster_num[index] += 1
        new_centroid[index] += item[0]
    return new_centroid / cluster_num
```

```python
def kmeans(myX,initial_centroids,n_iter,K):
    centroids = initial_centroids
    for x_iter in xrange(n_iter):
        idx = findClosestCentroids(myX,centroids,K)
        centroids = computeCentroids(myX,idx,K)
    return idx,centroids
    
```

```python
idx,centroids = kmeans(X,initial_centroids,10,3)
#print centroids.shape
#print centroids[1,:]
```

```python
def get_slice(X,idx,K):
    X_Slice = [[] for _ in range(K)]
    for i in xrange(idx.shape[0]):
        item = idx[i]
        index = int(item[0])
        #print index
        X_Slice[index].append(X[i])
    return X_Slice
```

```python
#print X_Slice[0][0] == X_Slice[0][1]
def slice_X_y(X_Slice_item):
    tempx,tempy = [],[]
    for item in X_Slice_item:
        tempx.append(item[0])
        tempy.append(item[1])
    return tempx,tempy
```

```python
def plotData(X_Slice,centroids):
    scatter_3 = go.Scatter(
        x = slice_X_y(X_Slice[0])[0],
        y = slice_X_y(X_Slice[0])[1],
        mode = 'markers',
        marker = dict(symbol="star",size=8),
        name = 'cluster 1'
    )
    scatter_4 = go.Scatter(
        x = slice_X_y(X_Slice[1])[0],
        y = slice_X_y(X_Slice[1])[1],
        mode = 'markers',
        marker = dict(symbol="cross",size=8),
        name = 'cluster 2'
    )
    scatter_5 = go.Scatter(
        x = slice_X_y(X_Slice[2])[0],
        y = slice_X_y(X_Slice[2])[1],
        mode = 'markers',
        marker = dict(symbol="circle",size=8),
        name = 'cluster 3'
    )
    scatter_6 =go.Scatter(
        x = centroids[:,0],
        y = centroids[:,1],
        mode = 'markers',
        marker = dict(symbol="star-diamond",size=12),
        name = 'centroid'
    )

    data = [scatter_3,scatter_4,scatter_5,scatter_6]
    return py.iplot(data)
    
```

```python
#from IPython.display import Image,display
X_Slice = get_slice(X,idx,K)
plotData(X_Slice,centroids)
```

### random initialize

```python
'''
we got a good result just because we have a good init_centroid in advance

if we randomly choose the centroid, what result would we get?
'''
```

```python
def chooseKRandomCentroids(myX, K):
    rand_indices = sample(xrange(0,myX.shape[0]),K)
    return np.array([myX[i] for i in rand_indices])
```
